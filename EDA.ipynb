# Analisis Exploratorio de datos (EDA) - Habitos de Estudiantes VS. Rendimiento

```python
# Importacion de librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carga de datos
df = pd.read_csv("student_habits_performance.csv")
```

```python
# Primeras lineas
print(display(df))
```

```python
# Informacion general del dataset (tipos de datos, valores nulos...)
print(df.info())
```

```python
# Estadisticas Descriptivas
print(df.describe())
```

## Hallazgos Iniciales
- Variables numericas (edad, horas de estudio, horas en redes, horas en netflix, porcentaje de asistencia, horas de sueño, clasificacion de salud mental, puntaje en el examen)
- Variables categoricas (id del estudiante, genero, trabajo de medio tiempo, calidad de dieta, nivel de educacion parental, participacion extracurricular)
- Se encuentra un valor inconsistente en las horas de estudio el cual es de 0
- Hay datos faltantes en educacion parental ( Podria indicar que los padres no terminaron bachillerato o secundaria)

## Analisis Univariado

```python
# Distribucion del puntaje del examen
plt.figure(figsize=(10,5))
sns.histplot(df['exam_score'], bins= 30, kde= True)
plt.title("Distribucion de puntajes en Examenes")
plt.xlabel('Puntaje')
plt.ylabel('Frecuencia')
plt.show()
```

- Se encuentra una distribucion sesgada a la derecha
- No necesariamente se encuentran valores atipicos en este punto del analisis ya que los valores bajos o altos dependen de los habitos. Haciendo referencia a la horas de estudio de valor 0, si la persona con este valor obtuvo un buen resultado, si se podria considerar un outlier o valor atipico 

```python
# Distribucion de horas de estudio
plt.figure(figsize=(12, 4))
sns.histplot(df['study_hours_per_day'], bins=20, kde=True)
plt.title("Horas de Estudio Diarias")
plt.show()
```

- La mayoria de los estudiantes estudia menos de 5 horas al dia


## Analisis bivariado


```python
# Correlacion entre Variables numericas
corr_matrix = df.select_dtypes(include=['float64', 'int64']).corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title("Matriz de Correlación")
plt.show()
```

- Las horas de estudio y el resultado del examen tiene una correlacion positiva
- Las horas en redes sociales y el resultado del examen tiene una correlacion negativa
- Las horas de sueño tienden hacia una correlacion positiva pero observando la grafica, estas no influyen en gran parte al resultado del examen

```python
# Rendimiento por genero
plt.figure(figsize=(8, 5))
sns.boxplot(x='gender', y='exam_score', data=df)
plt.title("Distribución de Puntajes por Género")
plt.show()
```

- No se encuentran diferencias significativas en el rendimiento entre generos
- Se encuentran valores atipicos tanto en hombres como en mujeres

## Analisis de segmentos

```python
# Crear categorías de rendimiento
df['performance_group'] = pd.cut(df['exam_score'], 
                                bins=[0, 50, 75, 100], 
                                labels=['Bajo', 'Medio', 'Alto'])

# Comparar hábitos entre grupos
group_stats = df.groupby('performance_group')[['study_hours_per_day', 'social_media_hours', 'sleep_hours']].mean()
print(group_stats)
```

- Los estudiantes con alto rendimiento se diferencian de los otros por la cantidad de horas que estudian
- La diferencia con los otros habitos es de decimas. Si tienen influencia pero no en gran medida
